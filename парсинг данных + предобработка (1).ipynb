```python
## преобразование ячеек в числовой формат
df=df.astype(float)
df
```

Пропусков в наших данных нет, поэтому можем переходить дальше. Мы считаем, что необходимо построить графики динамики курсов валют и ключевой ставки, чтобы увидеть, что с данными все более или менее в порядке, по курсам валют также можно увидеть исторические тенденции, кризисы и влияние каких-либо внешних шоков в соответствующие года

```python
def dynamic_graph(column):
    column=str(column)
    # the name of the currency
    df[column].plot()
    plt.title(column) # вывожу название компании из датафрейма
    plt.ylabel('Обратный курс к рублю')
    plt.xlabel('Дата')
    plt.show();
```

```python
dynamic_graph('Доллар США'), dynamic_graph('Евро'), dynamic_graph('Фунт стерлингов'), dynamic_graph('Китайский юань'), dynamic_graph('Японская йена'), dynamic_graph('Ключевая ставка')
```

По графикам динамики курсов валют можно увидеть, что Доллар США и Евро ведут себя практически одинаково, самой волатильной валютой за прошедшие годы являлся Китайский юань, а ключевая ставка в недавнее время очень сильно скакала.
