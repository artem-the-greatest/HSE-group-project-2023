```python
def get_rates(r):
    # получим доступ к сайту ЦБ
    url_USD = 'https://cbr.ru/currency_base/dynamics/?UniDbQuery.Posted=True&UniDbQuery.so=1&UniDbQuery.mode=1&UniDbQuery.date_req1=&UniDbQuery.date_req2=&UniDbQuery.VAL_NM_RQ={}.From=15.05.2015&UniDbQuery.To=15.05.2023'.format(r)
    otvet_USD = requests.get(url_USD, headers = {'User-Agent': UserAgent().chrome})

    # распарсим страничку в дерево 
    derevo_USD = BeautifulSoup(otvet_USD.content, 'html.parser')

    # найдём нужную часть информации сайта
    sth_USD = derevo.find_all('td')

    # сделаем список с курсом USD
    rates_USD = []
    dates_USD = []

    for i in range(3, len(sth_USD), 3):
        rates_USD.append(sth_USD[i].text)

    for i in range(1, len(sth_USD), 3):    
        dates_USD.append(sth_USD[i].text)
    
rates_USD[:10]
```
