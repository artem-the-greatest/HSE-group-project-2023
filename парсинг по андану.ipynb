# Данные процентных ставок

```python
### данные по ключевой ставке в России

# получим доступ к сайту ЦБ
url_key = 'https://cbr.ru/hd_base/KeyRate/?UniDbQuery.Posted=True&UniDbQuery.From=15.05.2015&UniDbQuery.To=15.05.2023'
otvet_key = requests.get(url_key, headers = {'User-Agent': UserAgent().chrome})

# распарсим страничку в дерево 
derevo_key = BeautifulSoup(otvet_key.content, 'html.parser')

# найдём нужную часть информации сайта
sth_key = derevo_key.find_all('td')

# сделаем список с ключевой ставкой
rates_key = []
dates_key = []

for i in range(1, len(sth_key), 2):
    rates_key.append(sth_key[i].text)
    
for i in range(0, len(sth_key), 2):    
    dates_key.append(sth_key[i].text)
    
print(rates_key[:10], dates_key[:10])
```

```python
value_RUB = pd.DataFrame({'Date': dates_key, 
                          'value_RUB': rates_key})
value_RUB
```

```python
### данные по LIBOR JPY

value_JPY = pd.read_csv("C:/Users/пк/Downloads/LIBOR JPY.csv")[['Date', '1M']]
value_JPY.rename(columns = {'1M': 'value_JPY'}, inplace = True)
value_JPY
```

```python
### данные по CNY

value_CNY = pd.read_csv("SHIBOR Overnight Bond Yield Historical Data.csv")[['Date', 'Price']]
value_CNY.rename(columns = {'Price': 'value_CNY'}, inplace = True)
value_CNY['Date'] = value_CNY['Date'].str.replace('/','.')

def func(date):
    date = date.split('.')
    date[0], date[1] = date[1], date[0]
    date = '.'.join(date)
    return(date)
    
value_CNY['Date'] = value_CNY['Date'].apply(lambda x: func(x))

value_CNY
```

# Соединим все датасеты

```python
# соединим датасеты с процентными ставками 

df_value = functools.reduce(lambda  left, right: 
                            pd.merge(left, right, on = ['Date'], how = 'outer'), 
                            [value_RUB, value_USD, value_EUR, value_GBP, value_JPY, value_CNY])
df_value
```

```python
# создадим общий датасет со всеми данными

df = pd.merge(rates, df_value, left_on = 'Дата', right_on = 'Date', how = 'inner')
df.drop(columns = ['Date'], inplace = True)
df
```
